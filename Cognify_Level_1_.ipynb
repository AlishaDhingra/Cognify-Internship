```python
top_cuisines = df.groupby("cuisine").size().reset_index().rename(columns={0:'count'}).sort_values("count",ascending=False).head(4)
sns.set(rc={'figure.figsize':(8,5)})
x = sns.barplot(x="cuisine",y="count",data=top_cuisines)
for bars in x.containers:
    x.bar_label(bars)
plt.title("Top Cuisine",fontsize=15)
plt.show()
```

# From the visualization we can conclude that top cuisines are
1. North Indian
2. North Indian and Chinese
3. Chinese and Fast Food together with same count.

```python
top_cuisine_rest = df[df['cuisine']=='North Indian'].groupby("name").size().reset_index().rename(columns={0:'count'}).sort_values("count",ascending=False).count()
per_top_1 = top_cuisine_rest['count']/df.shape[0]
per_top_1.round(3)
```

So only **0.029**% restaurants serve "North Indian" Cuisine

```python
top_cuisine_rest = df[df['cuisine']=='North Indian,Chinese'].groupby("name").size().reset_index().rename(columns={0:'count'}).sort_values("count",ascending=False).count()
per_top_2= top_cuisine_rest['count']/df.shape[0]
rounded_percentage = round(per_top_2, 3)
rounded_percentage
```

So only **0.034**% restaurants serve North Indian, Chinese Cuisine

```python
top_cuisine_rest = df[df['cuisine']=='Chinese'].groupby("name").size().reset_index().rename(columns={0:'count'}).sort_values("count",ascending=False).count()
per_top_3= top_cuisine_rest['count']/df.shape[0]
per_top_3.round(3)
```

So only **0.024**% restraunts serve Chinese Cuisine

# **Task3 : Price Range Distribution**
1. Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
2. Calculate the percentage of restaurants in each price range category.

```python
df['cost'] = df['cost'].astype(str).str.replace('₹ ', '').str.replace(',', '').astype(float)
```

```python
rest_cost_range = df.groupby('name', as_index=False).agg({'cost': 'mean'}).sort_values("cost", ascending=False)
cost_range_counts = rest_cost_range['cost'].value_counts()
cost_range_percentage = ((cost_range_counts / cost_range_counts.sum()) * 100).reset_index().rename(columns={"index": "cost", "cost_range": "percentage"}).sort_values(by='cost')
cost_range_percentage['cost'] = cost_range_percentage['cost'].astype("int")
cost_range_percentage = cost_range_percentage.groupby('cost', as_index=False).sum().reset_index(drop=True)
cost_range_percentage
```

```python
x= sns.barplot(x= cost_range_percentage['cost'],y= cost_range_percentage['count'])
for bars in x.containers:
    x.bar_label(bars)
plt.title("Percentage of Restaurants Price Range")
plt.show()
```
